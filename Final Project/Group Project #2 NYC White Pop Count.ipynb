```python
# First, we need to import commands to visualize our data:

# for general data wrangling tasks
import pandas as pd

# to read and visualize spatial data
import geopandas as gpd

# to provide basemaps 
import contextily as ctx

# to give more power to your figures (plots)
import matplotlib.pyplot as plt
```

```python
# Import population file
whitepop = gpd.read_file('ny.geojson')
```

```python
# How many rows and columns of data?
whitepop.shape
```

```python
# What does the data look like?
whitepop.head()
```

```python
# Let's visualize the dataset
whitepop.plot(figsize=(10,10))
```

```python
# Remove the county data row
whitepop = whitepop.drop([0])
```

```python
whitepop.head()
```

```python
# Clean the data to remove error columns
list(whitepop)
```

```python
columns_to_keep = ['geoid',
     'name',
 'B02001001',
 'B02001002',
 'B02001003',
 'B02001004',
 'B02001005',
 'B02001006',
 'B02001007',
 'B02001008',
 'B02001009',
 'B02001010',
 'geometry']              
```

```python
whitepop = whitepop[columns_to_keep]
```

```python
# Now we can visualize the cleaned up data
whitepop.head()
```

```python
# Rename headers of columns
whitepop.columns = ['geoid',
 'name',
 'Total',
 'White',
 'Black or African American',
 'American Indian and Alaska Native',
 'Asian',
 'Native Hawaiian and Other Pacific Islander',
 'Some other race alone',
 'Two or more races',
 'Two races including Some other race',
 'Two races excluding Some other race, and three or more races',
 'geometry']
```

```python
whitepop['White'].describe()
```

```python
plt.xlim(0, 15000)
whitepop['White'].plot.hist(bins=1000)
```

```python
whitepop_sorted = whitepop.sort_values(by='White',ascending = False)
```

```python
whitepop_sorted[['geoid','White']].describe()
```

```python
whitepop_sorted.head(100).plot()
```

```python
whitepop_sorted.head(100).plot(figsize=(10,10),column='White',legend=True)
```
